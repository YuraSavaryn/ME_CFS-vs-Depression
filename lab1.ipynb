```python
text_cols = dataset.select_dtypes(include='object').columns
num_cols = dataset.select_dtypes(include=np.number).columns

print("\nВиявлення аномалій")
for col in num_cols:
    Q1 = dataset[col].quantile(0.25)
    Q3 = dataset[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = dataset[(dataset[col] < Q1 - 1.5*IQR) | (dataset[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {outliers.shape[0]} аномальних значень")

    plt.figure(figsize=(8,4))
    sns.boxplot(x=dataset[col])
    plt.title(f"Аутлаєри у {col}")
    plt.show()
```

```python
print("\nРезультат простих агрегацій")
for col in text_cols[:3]:
    print(f"\nАгрегація за {col}:")
    print(dataset.groupby(col)[num_cols].mean().head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("D:\\ML\\university\\data\\me_cfs_vs_depression_dataset.csv")
```

```python
missing_data = pd.DataFrame({
    'dtype': dataset.dtypes,
    'missing': dataset.isna().sum(),
    'missing_dep': dataset[dataset['diagnosis'] == 'Depression'].isna().sum(),
    'missing_mecf': dataset[dataset['diagnosis'] == 'ME/CFS'].isna().sum(),
    'missing_both': dataset[dataset['diagnosis'] == 'Both'].isna().sum()
})
missing_data
```

```python
dataset_depression = dataset[dataset['diagnosis'] == 'Depression']
dataset_mecf = dataset[dataset['diagnosis'] == 'ME/CFS']
dataset_both = dataset[dataset['diagnosis'] == 'Both']

num_features = ['sleep_quality_index', 'brain_fog_level', 'physical_pain_score', 'stress_level', 'depression_phq9_score', 
            'fatigue_severity_scale_score', 'pem_duration_hours', 'hours_of_sleep_per_night']
categorical_features = ['gender', 'work_status', 'social_activity_level', 'exercise_frequency', 'meditation_or_mindfulness']

for feature in num_features:
    dataset_depression[feature].fillna(dataset_depression[feature].mean(), inplace=True)
    dataset_mecf[feature].fillna(dataset_mecf[feature].mean(), inplace=True)
    dataset_both[feature].fillna(dataset_both[feature].mean(), inplace=True)

for feature in categorical_features:
    dataset_depression[feature].fillna(dataset_depression[feature].mode()[0], inplace=True)
    dataset_mecf[feature].fillna(dataset_mecf[feature].mode()[0], inplace=True)
    dataset_both[feature].fillna(dataset_both[feature].mode()[0], inplace=True)

dataset.update(dataset_depression)
dataset.update(dataset_mecf)
dataset.update(dataset_both)
```

```python
missing_data = pd.DataFrame({
    'dtype': dataset.dtypes,
    'missing': dataset.isna().sum(),
    'missing_dep': dataset[dataset['diagnosis'] == 'Depression'].isna().sum(),
    'missing_mecf': dataset[dataset['diagnosis'] == 'ME/CFS'].isna().sum(),
    'missing_both': dataset[dataset['diagnosis'] == 'Both'].isna().sum()
})
missing_data
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset['diagnosis'] = le.fit_transform(dataset['diagnosis'])

categorical_encoding = pd.get_dummies(dataset[categorical_features], columns=categorical_features, drop_first=True)
dataset = pd.concat([dataset.drop(columns=categorical_features), categorical_encoding], axis=1)
```

```python
Q1 = dataset['depression_phq9_score'].quantile(0.25)
Q3 = dataset['depression_phq9_score'].quantile(0.75)
IQR = Q3 - Q1

outliers1 = dataset[(dataset['depression_phq9_score'] < Q1 - 1.5*IQR) | (dataset['depression_phq9_score'] > Q3 + 1.5*IQR)]
outliers1.shape
```

```python
dataset.drop(index=outliers1.index, inplace=True)
dataset.shape
```

```python
Q1 = dataset['fatigue_severity_scale_score'].quantile(0.25)
Q3 = dataset['fatigue_severity_scale_score'].quantile(0.75)
IQR = Q3 - Q1

outliers2 = dataset[(dataset['fatigue_severity_scale_score'] < Q1 - 1.5*IQR) | (dataset['fatigue_severity_scale_score'] > Q3 + 1.5*IQR)]
outliers2.shape
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
dataset['fatigue_severity_scale_score'] = np.clip(dataset['fatigue_severity_scale_score'], lower, upper)
```

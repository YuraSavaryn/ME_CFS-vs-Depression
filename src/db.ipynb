```python
import os
from dotenv import load_dotenv
import numpy as np
import pandas as pd
from sqlalchemy import create_engine

load_dotenv()

dataset = pd.read_csv(os.getenv("DATASET_URL"))

engine = create_engine(os.getenv("DATABASE_URL"))

dates = pd.date_range('2025-01-01', '2025-12-31')
dataset['timestamp'] = np.random.choice(dates, 1000)

dataset.to_sql("data", engine, if_exists="replace", index=True, index_label='data_id')
```

```python
import os
from dotenv import load_dotenv
import numpy as np
import pandas as pd
from sqlalchemy import create_engine

load_dotenv()

dataset = pd.read_csv(os.getenv("CLEAN_DATASET_URL"))

engine = create_engine(os.getenv("DATABASE_URL"))

dates = pd.date_range('2025-01-01', '2025-12-31')
dataset['timestamp'] = np.random.choice(dates, dataset.shape[0])

dataset.to_sql("clean_data", engine, if_exists="replace", index=True, index_label='data_id')
```
